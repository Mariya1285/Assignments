```python
import pandas as pd

df = pd.read_csv("takehome_user_engagement-intern.csv")
count_df = df.groupby(['user_id']).count()
print(count_df)
count_df = count_df.reset_index()
print(count_df)
# print(type(count_df['visited']))
user_id_list = list()
exclude = list()
for ind, row in count_df.iterrows():
  if row['visited']>3:
    user_id_list.append(row['user_id'])
  else:
    exclude.append(row['user_id'])
print(user_id_list)

# take_home_users = ([lambda x: x['object_id'] in user_id_list] pd.read_csv("takehome_users-intern.csv"))
take_home_users =pd.read_csv("takehome_users-intern.csv", encoding='latin-1')

mask = take_home_users['object_id'].isin(user_id_list)
adopted_user = take_home_users[mask]
adopted_user['label'] = "A"

normal_user = take_home_users[~mask]
normal_user['label'] = "B"
print(take_home_users+normal_user)
```
